```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
# andom data
x_2 = np.linspace(-2, 2, 1000)

#g(x)
def g(x):
    return x**4 - 4*x**2 + 5

def dg(x):
    return 4*x**3 - 8*x
```

```python
def gradient_descent(derivitive_func, initial_guess, multiplier = 0.02, precision = 0.001):
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivitive_func(new_x)]

    for i in range (1300):
        old_x = new_x
        gradient = derivitive_func(old_x)
        new_x = old_x - multiplier * gradient    

        step_size = abs(new_x - old_x)        
        x_list.append(new_x)
        slope_list.append(derivitive_func(new_x))

        #print(step_size)


        if step_size < precision:
            print("Precision met at loop number:", i)
            break
    return new_x, x_list, slope_list
    
```

```python
# Callin gradient_descent
local_min, list_x, deriv_list = gradient_descent(derivitive_func = dg, initial_guess = 1.5)
plt.figure(figsize=[16,5])

# Chart: Cost function
plt.subplot(1, 2, 1)
plt.plot(x_2, g(x_2), alpha=0.6)
plt.scatter(x_list, g(np.array(x_list)), color="red", s=100, alpha=0.4)

#style plot
plt.xlim(-2,2)
plt.ylim(0.5,5.5)
plt.title('Cost function', fontsize=17)
plt.xlabel('x', fontsize=16)
plt.ylabel('g(x)', fontsize=16)
plt.grid()


# Chart derivative 
plt.subplot(1, 2, 2)
plt.plot(x_2, dg(x_2), c='skyblue', linewidth=5, alpha=0.4)



#style plot
plt.xlim(-2,2)
plt.ylim(-6,8)
plt.title('Slope of the Cost function', fontsize=17)
plt.xlabel('x', fontsize=16)
plt.ylabel('dg(x)', fontsize=16)
plt.grid()


plt.show()
```
