# Imports

````python
import numpy as np
import tensorflow as tf
from tensorflow.compat.v1.saved_model import load, tag_constants
````

# Constants

````python
X_TEST_PATH = 'C:/Users/BediZ/DS-ML/Jupyter Notebook - ML Projects/resources/MNIST/load_xtest.csv'
Y_TEST_PATH = 'C:/Users/BediZ/DS-ML/Jupyter Notebook - ML Projects/resources/MNIST/load_ytest.csv'
EXPORT_MODEL = 'C:/Users/BediZ/DS-ML/Jupyter Notebook - ML Projects/05 Neural Nets - Retrainer Image Classification/SavedModel'
````

# Load the Data

````python
y_test = np.loadtxt(Y_TEST_PATH, delimiter=',', dtype=int)
print(f'Shape of y_test {y_test.shape}')
````

````python
x_test = np.loadtxt(X_TEST_PATH, delimiter=',', dtype=float)
print(f'Shape of x_test {x_test.shape}')
````

````python
np.savetxt(fname='x_test0_ylabel7.txt', X=[x_test[0]], delimiter=', ', fmt='%1.2f')
````

# Create Session

````python
graph = tf.Graph()
session_object = tf.compat.v1.Session(graph=graph)
````

# Load the Model

```
tf.compat.v1.saved_model.load(sess,
                           [tag_constants.SERVING],
                           EXPORT_DIR)
```

````python

load(sess=session_object, tags=[tag_constants.SERVING], export_dir=EXPORT_MODEL)
````

# Run Session & Make Prediction

````python
X = graph.get_tensor_by_name('X:0') 
# 0 is the first output index of X in the graph in TensorBoard -> graph -> input (in this case it is X)
y_prediction = graph.get_tensor_by_name('accuracy_calculation/prediction:0') 
````

````python
prediction = session_object.run(fetches=y_prediction, feed_dict={X: x_test})
````

# Check Results

````python
y_test[:5]
````

````python
prediction[:5]
````

````python
number_correct_predictions = sum(np.equal(prediction, y_test))
number_correct_predictions / len(prediction)
````

# Cleanup

````python
session_object.close()
tf.compat.v1.reset_default_graph()
````
