### The Learning Rate

```python
import matplotlib.pyplot as plt
import numpy as np
import sys

%matplotlib inline
```

```python
# Make data
x_3 = np.linspace(start= -2.5, stop=2.5, num=100)

def h(x):
    return x**5 - 2 * x**4 + 2

def dh(x):
    return 5 * x**4 - 8 * x**3
```

```python
def gradient_descent(derivitive_func, initial_guess, multiplier = 0.02, precision = 0.001, max_i=300):
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivitive_func(new_x)]

    for i in range (max_i):
        old_x = new_x
        gradient = derivitive_func(old_x)
        new_x = old_x - multiplier * gradient    

        step_size = abs(new_x - old_x)        
        x_list.append(new_x)
        slope_list.append(derivitive_func(new_x))

        #print(step_size)


        if step_size < precision:
            print("Local min occures at(x axis): ", new_x)
            print("Cost at this minimum is(y axis): ", h(new_x))
            print("Precision met at loop number:", i)
            break
    return new_x, x_list, slope_list
    
```

```python
# Callin gradient_descent
inter_num = 30
low_gamma = local_min, list_x, deriv_list = gradient_descent(derivitive_func = dh, 
                                                 initial_guess = 1.9, 
                                                 multiplier = 0.0005, 
                                                 precision = 0.0001,
                                                 max_i=inter_num)
mid_gamma = local_min, list_x, deriv_list = gradient_descent(derivitive_func = dh, 
                                                 initial_guess = 1.9, 
                                                 multiplier = 0.001, 
                                                 precision = 0.0001,
                                                 max_i=inter_num)
high_gamma = local_min, list_x, deriv_list = gradient_descent(derivitive_func = dh, 
                                                 initial_guess = 1.9, 
                                                 multiplier = 0.002, 
                                                 precision = 0.0001,
                                                 max_i=inter_num)
ultra_high_gamma = local_min, list_x, deriv_list = gradient_descent(derivitive_func = dh, 
                                                 initial_guess = 1.9, 
                                                 multiplier = 0.022, 
                                                 precision = 0.001,
                                                 max_i=inter_num)

# Plotting reduction in cost for each iteration
plt.figure(figsize=[20,10])
#style plot
plt.xlim(0,40)
plt.ylim(-1,1)
plt.title('Effecti of Learning', fontsize=17)
plt.xlabel('Number of itterations', fontsize=16)
plt.ylabel('Cost', fontsize=16)
plt.grid()

# Values of the chart
## X Axis Data: convert the list to numpy arrays
low_values = np.array(low_gamma[1])
mid_values = np.array(mid_gamma[1])
high_values = np.array(high_gamma[1])
ultra_high_values = np.array(ultra_high_gamma[1])

## X Axis Data:  create the list form 0 ti n+1
iteration_list= list(range(0, inter_num+1))

plt.plot(iteration_list, h(low_values), color="green", alpha=0.6) 
plt.scatter(iteration_list, h(low_values), color="green", s=100, alpha=0.4)

plt.plot(iteration_list, h(mid_values), color="blue", alpha=0.6) 
plt.scatter(iteration_list, h(mid_values), color="blue", s=100, alpha=0.4)

plt.plot(iteration_list, h(high_values), color="yellow", alpha=0.6) 
plt.scatter(iteration_list, h(high_values), color="yellow", s=100, alpha=0.4)

plt.plot(iteration_list, h(ultra_high_values), color="red", alpha=0.6) 
plt.scatter(iteration_list, h(ultra_high_values), color="red", s=100, alpha=0.4)




plt.show()
```
