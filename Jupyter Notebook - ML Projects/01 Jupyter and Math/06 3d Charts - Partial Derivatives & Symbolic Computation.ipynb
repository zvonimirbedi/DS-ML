### 3d Charts

```python
import matplotlib.pyplot as plt
import numpy as np
import sys

from mpl_toolkits.mplot3d.axes3d import Axes3D
from matplotlib import cm #colores for the 3d chart depth
from sympy import symbols, diff


%matplotlib inline
```

```python
x_4 = np.linspace(start= -2, stop=2, num=55)
y_4 = np.linspace(start= -2, stop=2, num=55)

print("Array: \n", x_4, "\n")
x_4, y_4 = np.meshgrid(x_4, y_4)
print("3d: \n", x_4)
```

```python
# Make datax_3 = np.linspace(start= -2.5, stop=2.5, num=100)

def f(x, y):
    r=3**(-x**2 - y**2)
    return 1/(r + 1)

```

```python
fig = plt.figure(figsize=[16,12])
ax=fig.gca(projection='3d')

ax.set_xlabel("X vertex", fontsize=20)
ax.set_ylabel("Y vertex", fontsize=20)
ax.set_zlabel("Cost - Z vertex", fontsize=20)


ax.plot_surface(x_4, y_4, f(x_4, y_4), cmap=cm.coolwarm_r , alpha=0.7)
plt.show()
#help(cm)
```

### Partial Derivatives & Symbolic Computation

```python
a, b = symbols('x, y')
f(a, b)
print('Our Cost function f(x, y) is: ', f(a, b))
print('Partial Derivative with respect to x is: ', diff(f(a, b), a))
print('Value of f(x, y) at x=1.8, y=1.0', 
      f(a, b).evalf(subs={a:1.8, b:1.0})) # Python Dictionary
print('Value of partial derivative with respect to x: ',diff(f(a, b), a).evalf(subs={a:1.8, b:1.0}))
```
