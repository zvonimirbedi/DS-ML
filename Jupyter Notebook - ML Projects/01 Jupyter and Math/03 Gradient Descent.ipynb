### Example 1 - a simple cost function: 
#### $$f(x) = x^2 + x + 1$$

```python
def f(x):
    return x**2 + x + 2
```

```python
# linspace - start number, end number, output of between numbers with equal distances 
x_1 = np.linspace(start=-3, stop=3, num=500)
#print(x_1)
```

```python
# Plot the linespace data
plt.plot(x_1, f(x_1))

#style plot
plt.xlim(-4,4)
plt.ylim(0,15)
plt.xlabel('x', fontsize=16)
plt.ylabel('f(x)', fontsize=16)

plt.show()
```

### Slope and Derivaties
Function for the derivateive od f(x) called df(x)

```python
def df(x):
    return 2 * x + 1
```

```python
# Plot the linespace data and derivative side by side

plt.figure(figsize=[16,5])

# Chart: Cost function
plt.subplot(1, 2, 1)
plt.plot(x_1, f(x_1))

#style plot
plt.xlim(-4,4)
plt.ylim(0,15)
plt.title('Cost function', fontsize=17)
plt.xlabel('x', fontsize=16)
plt.ylabel('f(x)', fontsize=16)
plt.grid()


# Chartderivative 
plt.subplot(1, 2, 2)
plt.plot(x_1, df(x_1), c='skyblue', linewidth=5)

#style plot
plt.xlim(-4,4)
plt.ylim(-6,8)
plt.title('Slope of the Cost function', fontsize=17)
plt.xlabel('x', fontsize=16)
plt.ylabel('df(x)', fontsize=16)
plt.grid()


plt.show()
```

```python
# Gradient Descent
new_x = 2
old_x = 0
step_multiplier = 0.1
precision = 0.0001

x_list = [new_x]
slope_list = [df(new_x)]

for i in range (130):
    old_x = new_x
    gradient = df(old_x)
    new_x = old_x - step_multiplier * gradient    
    
    x_list.append(new_x)
    slope_list.append(df(new_x))
    
    step_size = abs(new_x - old_x)
    #print(step_size)
    
    
    if step_size < precision:
        print("Precision met at loop number:", i)
        break

print('Local minimum occurs at:', new_x)
print('Slope or df(x) value at this point is:', df(new_x))
print('Cost or f(x) value at this point is:', f(new_x))

    
```

```python
# Superimpose gradient descent calculations on plot

plt.figure(figsize=[16,5])

# Chart: Cost function
plt.subplot(1, 2, 1)
plt.plot(x_1, f(x_1), alpha=0.6)

x_array = np.array(x_list)
plt.scatter(x_list, f(x_array), color="red", s=100, alpha=0.4)

#style plot
plt.xlim(-4,4)
plt.ylim(0,15)
plt.title('Cost function', fontsize=17)
plt.xlabel('x', fontsize=16)
plt.ylabel('f(x)', fontsize=16)
plt.grid()


# Chart derivative 
plt.subplot(1, 2, 2)
plt.plot(x_1, df(x_1), c='skyblue', linewidth=5, alpha=0.4)

x_array = np.array(x_list)
plt.scatter(x_list, slope_list, color="red", s=100, alpha=0.6)


#style plot
plt.xlim(-4,4)
plt.ylim(-6,8)
plt.title('Slope of the Cost function', fontsize=17)
plt.xlabel('x', fontsize=16)
plt.ylabel('df(x)', fontsize=16)
plt.grid()


plt.show()
```

### Gradient Descent as a Python Function

```python
def gradient_descent(derivitive_func, initial_guess, multiplier = 0.02, precision = 0.001):
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivitive_func(new_x)]

    for i in range (1300):
        old_x = new_x
        gradient = derivitive_func(old_x)
        new_x = old_x - multiplier * gradient    

        step_size = abs(new_x - old_x)        
        x_list.append(new_x)
        slope_list.append(derivitive_func(new_x))

        #print(step_size)


        if step_size < precision:
            print("Precision met at loop number:", i)
            break
    return new_x, x_list, slope_list
    
```

```python
local_min, list_x, deriv_list = gradient_descent(derivitive_func = dg, 
                                                 initial_guess = 0.5, 
                                                 multiplier = 0.01, 
                                                 precision = 0.0001)
print('Local minimum occurs at:', local_min)
print('Number of loops:', len(list_x))
```

```python
local_min, list_x, deriv_list = gradient_descent(derivitive_func = dg, 
                                                 initial_guess = 0.1)
print('Local minimum occurs at:', local_min)
print('Number of loops:', len(list_x))
```
