### Example 3 - Divergence, Overflow and Python Tuples
#### $$ h(x) = x^5 - 2x^4 + 2 $$

```python
# Make data
x_3 = np.linspace(start= -2.5, stop=2.5, num=100)

def h(x):
    return x**5 - 2 * x**4 + 2

def dh(x):
    return 5 * x**4 - 8 * x**3
```

```python
def gradient_descent(derivitive_func, initial_guess, multiplier = 0.02, precision = 0.001, max_i=300):
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivitive_func(new_x)]

    for i in range (max_i):
        old_x = new_x
        gradient = derivitive_func(old_x)
        new_x = old_x - multiplier * gradient    

        step_size = abs(new_x - old_x)        
        x_list.append(new_x)
        slope_list.append(derivitive_func(new_x))

        #print(step_size)


        if step_size < precision:
            print("Local min occures at(x axis): ", new_x)
            print("Cost at this minimum is(y axis): ", h(new_x))
            print("Precision met at loop number:", i)
            break
    return new_x, x_list, slope_list
    
```

```python
# Callin gradient_descent
local_min, list_x, deriv_list = gradient_descent(derivitive_func = dh, initial_guess = 0.2, max_i=300)
plt.figure(figsize=[16,5])

# Chart: Cost function
plt.subplot(1, 2, 1)
plt.plot(x_3, h(x_3), alpha=0.6)
 
plt.scatter(list_x, h(np.array(list_x)), color="red", s=100, alpha=0.4)

#style plot
plt.xlim(-2,2)
plt.ylim(-2,5)
plt.title('Cost function', fontsize=17)
plt.xlabel('x', fontsize=16)
plt.ylabel('h(x)', fontsize=16)
plt.grid()


# Chart derivative 
plt.subplot(1, 2, 2)
plt.plot(x_3, dh(x_3), c='skyblue', linewidth=5, alpha=0.4)

plt.scatter(list_x, dh(np.array(list_x)), color="red", s=100, alpha=0.4)



#style plot
plt.xlim(-2,2)
plt.ylim(-6,8)
plt.title('Slope of the Cost function', fontsize=17)
plt.xlabel('x', fontsize=16)
plt.ylabel('dh(x)', fontsize=16)
plt.grid()


plt.show()
```
