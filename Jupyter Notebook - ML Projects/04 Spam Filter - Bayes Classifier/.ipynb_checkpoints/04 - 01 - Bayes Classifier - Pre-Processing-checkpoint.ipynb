## Email Body Extraction

```python
def email_body_generator(path):
    for root, dirnames, filenames in walk(path):
        for file_name in filenames:
            
            filepath = join(root, file_name)
            stream = open(filepath, encoding='latin-1')

            is_body = False
            body_lines = []

            for line in stream:
                if is_body:
                    body_lines.append(line)
                elif line == '\n':
                    is_body = True       

            stream.close()

            email_body = '\n'.join(body_lines)

            yield file_name, email_body
```

```python
def data_frame_from_directory(path, classification):
    rows = []
    row_names = []
    
    for file_name, email_body in email_body_generator(path):
        rows.append({'MESSAGE': email_body, 'CATEGORY': classification})
        row_names.append(file_name)
        
    return pd.DataFrame(rows, index = row_names)
```

### Functions for Email Processing

```python
def clean_message(message, stemmer=PorterStemmer(), stop_words=set(stopwords.words('english'))):
    
    # Converte to lower case
    words = word_tokenize(message.lower())
    
    clean_words =[]
    
    for word in words:
        # Remove the stop words and punctuation
        if word not in stop_words and word.isalpha():
            clean_words.append(stemmer.stem(word))
    
    
    return clean_words
```

```python
clean_message(email_body)
```

```python
def clean_message_no_html(message, stemmer=PorterStemmer(), stop_words=set(stopwords.words('english'))):
    
    # Remove HTML tags
    message_no_html_soup = BeautifulSoup(message, 'html.parser')
    cleand_html_message = message_no_html_soup.get_text()
    
    # Converte to lower case
    words = word_tokenize(cleand_html_message.lower())
    
    clean_words = []
    
    for word in words:
        # Remove the stop words and punctuation
        if word not in stop_words and word.isalpha():
            clean_words.append(stemmer.stem(word))
    
    
    return clean_words
```

```python
clean_message_no_html(all_emails.at[2, 'MESSAGE'])
```

## Create a Sparse Matrix for the Training Data

```python
word_index = pd.Index(vocabulary.VOCABULARY_WORD)
type(word_index[3])
```

```python
word_index.get_loc('take')
```

```python
def make_sparse_matrix(dataframe, indexed_words, labels):
    """
    Returns sparsed matrix as dataframe
    dataframe: a dataframe with words in the columns with a document id as an index (X_train or X_test)
    indexed_words: index of words ordered by word id
    labels: category as a series (y_train or y_test)
    """
    
    number_of_rows = dataframe.shape[0]
    number_of_columns = dataframe.shape[1]
    word_set = set(indexed_words)
    dictionary_list = []
    
    for i in range(number_of_rows):
        for j in range(number_of_columns):
            word = dataframe.iat[i,j]
            if word in word_set:
                doc_id = dataframe.index[i]
                category = labels.at[doc_id]
                word_id = indexed_words.get_loc(word)
                item = {'LABEL': category, 'DOC_ID': doc_id, 'OCCURENCE': 1, 'WORD_ID': word_id}
                dictionary_list.append(item)
                
                
    return pd.DataFrame(dictionary_list)
```

```python
%%time

sparse_train_df = make_sparse_matrix(X_train, word_index, y_train)
```
